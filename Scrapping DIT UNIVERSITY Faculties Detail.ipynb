**Importing necessary liberaries**

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

***Let's setting up the neccessary variables and Send an HTTP GET request and retrieve the HTML content***

```python
url="https://www.dituniversity.edu.in/faculty"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
    'Accept-Language': 'en-US,en;q=0.9',
}
```

```python
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

**Find and extract the desired data using appropriate selectors**

```python
Faculties=soup.find_all("div",class_="col-md-3")
#we iterate through each faculty div element obtained from soup.find_all("div", class_="col-md-3").
```

```python
Faculties
```

```python
faculty_names = []
degrees = []
schools = []


for faculty in Faculties:
    try:
        faculty_name = faculty.find("a").text.strip() #We apply the text.strip() method to remove any leading or trailing whitespace
    except AttributeError:
        faculty_name = "N/A"  # Set a default value if faculty name is not found
    faculty_names.append(faculty_name)

    try:
        degree = faculty.find("p").find("span").text.strip()
    except AttributeError:
        degree = "N/A"  # Set a default value if degree is not found
    degrees.append(degree)

    try:
        school = faculty.find_all("p")[1].text.strip()
    except IndexError:
        school = "N/A"  # Set a default value if school is not found
    schools.append(school)
```

**Print or process the extracted data as required**


```python
for faculty_name, degree, school in zip(faculty_names, degrees, schools):
    print(f"Faculty Name: {faculty_name}")
    print(f"Degree: {degree}")
    print(f"School: {school}")
    print()
```

```python
# Create a dictionary using the extracted data
data = {
    'faculty_name': faculty_names,
    'degrees': degrees,
    'schools': schools
}

# Convert the dictionary into a pandas DataFrame
df = pd.DataFrame(data)
```
